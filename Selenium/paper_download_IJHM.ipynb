```python
for j in range(len(sd)):
    
    if j%200==0:
        sd.to_excel(f"Data/IJHM_{j}.xlsx")
    
    try:
        print("main", j, sd.iloc[j].urls)
        driver.get(sd.iloc[j].urls)
    except:
        try:
            print("except", j, sd.iloc[j].urls)

            time.sleep(5)
            driver.get(sd.iloc[j].urls)
        except:
            print(j, "fail")
            sd.loc[j, "filename"] = "No Url"
            continue
    
    doc = "" 
    time.sleep(random.randint(3,15))
    
    secs =driver.find_elements_by_xpath('//div[@class="Body u-font-serif"]/div/section')

    i=1
    for sec in secs:
        doc += f"***\nSECTION 00{str(i)}\n***"
        doc += "*****\n*****"
        doc += sec.text
        i+=1
        
    try:
        title = driver.find_element_by_xpath('//*[@id="screen-reader-main-title"]/span').text
        sd.loc[j, "TI"] = title
    except:
        sd.loc[j, "TI"] = ""
    
    if len(doc)>100:
        filename = f"IJHM_{j}.txt"
    else:
        filename = f"Pro_IJHM_{j}.txt"

    with io.open(f"Data/pap1/{filename}", "w", encoding="utf-8") as text_file:
        text_file.write(doc)
    
    sd.loc[j, "filename"] = filename
    time.sleep(1)
```
