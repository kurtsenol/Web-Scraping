```python
for j in range(len(sd)):
    try:
        print("main", j, sd.iloc[j].urls)
        driver.get(sd.iloc[j].urls)
    except:
        try:
            print("except", j, sd.iloc[j].urls)

            time.sleep(5)
            driver.get(sd.iloc[j].urls)
        except:
            print(j, "fail")
            sd.loc[j, "filename"] = "No Url"
            continue
    
    doc = "" 
    time.sleep(random.randint(2,15))
    
    secs = driver.find_elements_by_xpath('//div[@class="hlFld-Fulltext"]/div[contains(@id,"00")]')

    i=1
    for sec in secs:
        
        doc += f"***\nSECTION 00{str(i)}\n***"
        doc += "*****\n*****"
        doc += sec.text
        i+=1

#     try:
#         title = driver.find_element_by_xpath('//*[@id="25e8d4fe-c5d2-4686-bdef-3b125e01db4b"]/div/div/div[1]/div/div[1]/div').text
#         sd.loc[j, "TI"] = title
#     except:
#         sd.loc[j, "TI"] = ""
    
    if len(doc)>100:
        filename = f"JHMM_{j}.txt"
    else:
        filename = f"Pro_JHMM_{j}.txt"

    with io.open(f"Data/pap4/{filename}", "w", encoding="utf-8") as text_file:
        text_file.write(doc)
    
    sd.loc[j, "filename"] = filename
    time.sleep(1)
```
